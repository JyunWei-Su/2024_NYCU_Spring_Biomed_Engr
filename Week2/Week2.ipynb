(3c) Load the 'topography180x360grid.txt' file again. For all points with elevations above 2000 m, set the elevation to 10000 m. For all points with a pessimetry of more then 2000 m, set the elevation to -10000 m.

```python
H4 = np.loadtxt('topography_180x360_grid.txt')
for i in range(0,180):
    for j in range(0,360):
        if H4[i,j]>2000:
            H4[i,j] = 10000
        elif H4[i,j] < -2000:
            H4[i,j] = -10000
plt.imshow(H4, interpolation='nearest')
plt.show()
```

(3d) In addition to the modifications in (3c), now also change the elevation to zero if the original elevation was betweenb -2000 and +2000 m.

```python
H4 = np.loadtxt('topography_180x360_grid.txt')
for i in range(0,180):
    for j in range(0,360):
        if H4[i,j]>2000:
            H4[i,j] = 10000
        elif H4[i,j] < -2000:
            H4[i,j] = -10000
        else:
            H4[i,j] = 0
plt.imshow(H4, interpolation='nearest')
plt.show()
```
