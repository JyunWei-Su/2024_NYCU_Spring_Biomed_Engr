**Stationary State of the Heat Equation in 2D**

In this section, we want to find the stationary temperature distribution in a 2D square with the following boundary conditions:

image.png



Imagine you are renting a room of size $L\times L$ that has three cold outside walls and one wall that is
heated by your neighbor. The outside temperature is kept fixed at 0°C and inside wall at 100°C.
Divide the square into $N \times N$ segments. $N = 51$ is a good resolution to start. Solving this problem
numerically requires a 2D array $T_{i,j}$ or in Python language an array `T[i,j]`.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

N = 51
L = 1.0
T_outside = 100
T_inside = 0

X = np.linspace(0, L, N)
Y = np.linspace(0, L, N)
XX, YY = np.meshgrid(X, Y)

#set the boundary condition
T = np.zeros((N,N),dtype=float)
T_new = np.zeros((N,N),dtype=float)

T[:,0] = T_outside

iteration = 0
iteration_max = 100

while iteration < iteration_max:
    T_new[:,0] = T_outside
    for i in range(1,N-1):
        for j in range(1,N-1):
            T_new[i,j]=0.25*(T[i,j-1]+T[i-1,j]+T[i+1,j]+T[i,j+1])
    iteration += 1
    T = T_new.copy()

    if iteration % 20 == 0:
        fig = plt.figure(dpi=200)
        ax = fig.add_subplot(projection='3d')
        # ax.plot_surface(XX, YY, T_new, cmap=cm.coolwarm, antialiased=False)
        ax.plot_surface(XX, YY, T_new, cmap='jet', antialiased=False)
        plt.draw()
        plt.pause(0.05)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.animation import FFMpegWriter

N = 51
L = 1.0
T_outside = 100
T_inside = 0

X = np.linspace(0, L, N)
Y = np.linspace(0, L, N)
XX, YY = np.meshgrid(X, Y)

#set the boundary condition
T = np.zeros((N,N))
T_new = np.zeros((N,N))

T[:,0] = T_outside

iteration = 0
iteration_max = 300

fig = plt.figure(dpi = 200)
writer = FFMpegWriter(fps = 30, bitrate = 200000)

with writer.saving(fig, "heat.mp4", dpi = 200):
    while iteration < iteration_max:
        T_new[:,0] = T_outside
        for i in range(1,N-1):
            for j in range(1,N-1):
                T_new[i,j]=0.25*(T[i,j-1]+T[i-1,j]+T[i+1,j]+T[i,j+1])
        iteration += 1
        T = T_new.copy()

        fig.clear()
        ax = fig.add_subplot(projection = '3d')
        #ax.plot_surface(XX, YY, T, cmap = cm.coolwarm, antialiased = False)
        ax.plot_surface(XX, YY, T, cmap = 'jet', antialiased = False)
        plt.draw()
        plt.pause(0.01)
        writer.grab_frame()
```
