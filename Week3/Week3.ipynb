With the command `np.loadtxt(...)`, load the solution file and your data set. Store the length of the data sets in the variable `nNumbers`. We know the number is 100 but Jupyter does not. Use this variable whenever it is needed later. Plot your data set and the solutions in one diagram. Add a labels and legend with `plt.legend()`. Insert one comment how well they agree.

```python
PATH_SOLUTION = 'x_positions_2019.txt'
PATH_MINEDATA = '5018556.txt'
```

```python
solution = np.loadtxt(PATH_SOLUTION)
data = np.loadtxt(PATH_MINEDATA)

nNumbers = solution.shape[0]
```

```python
plt.rcParams['figure.figsize'] = [6,4]
plt.plot(solution,'o-',label='mine')
plt.plot(data,'^r-',label='sol')
plt.legend(loc='upper right')
plt.show()
```

```python
plt.rcParams['figure.figsize'] = [6,4]
plt.scatter(np.arange(nNumbers), solution, color='blue', label = 'mine')
plt.scatter(np.arange(nNumbers), data, color='red', label = 'sol')
```

Plot the difference between your data and the solution. (Taking the difference requires only one line of code.) Add a thin line that marks zero deviation. Print the values of the most positive and most negative deviation.
Please briefly explain these extreme descepancies.

```python
diff = data - solution
plt.rcParams['figure.figsize'] = [6,4]
plt.plot(diff, 'ko-', lw=2, ms=3, mfc='white')
plt.plot(diff*0.0, 'k--',lw=1, ms=2,mfc='white')
plt.show()
print('MAX:', max(diff), 'MIN:', min(diff))
```

Compute the average of the difference between your data and the solution. Include the sign of the difference in the summation. Then write a comment on the sign of the average.

```python
# avg = np.mean(diff)
avg = diff.sum()/nNumbers
print('AVG:', avg)
```

Compute and print the variance and the standard deviation sigma between your data set and the solution as described on the lecture notes.

```python
var = sum(np.power(diff - avg, 2))/nNumbers
std = np.sqrt(var)
print('VAR:', var)
print('STD:', std)
```

Let us assume we know the average is zero. Let us recalculate the variance and the standard deviation.

```python
avg = 0
var = sum(np.power(diff - avg, 2))/nNumbers
std = np.sqrt(var)
print('VAR:', var)
print('STD:', std)
```
